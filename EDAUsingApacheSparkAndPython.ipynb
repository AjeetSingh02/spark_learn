In last video we saw that in order to apply SQL on a dataframe we have to first save it as a table. 
Similarly to run py-spark on a table we have to convert it to dataframe.

Here's how we will do that

```
table_name = "intermediate_transformations"
lc_df = spark.table(table_name)
display(lc_df)
```

```
%sql

select addr_state, sum(loan_amnt) from intermediate_transformations group by addr_state

-- There is a lot of junk in this data. This data is uncleaned
```

```
# Doing the same in python but calculating count and not sum

from pyspark.sql.functions import isnan, when, count, col, log

display(lc_df.groupBy("addr_state").agg((count(col("annual_inc"))).alias("count")))
```
